```python
# checking the area with maximum transgender population both in rural and urban areas
max_trans_rural = df.loc[df['TRANSGENDER (RURAL)'].idxmax()]
max_trans_urban = df.loc[df['TRANSGENDER (URBAN)'].idxmax()]
print("Max Transgender (Rural):")
print(max_trans_rural)
print("\nMax Transgender (Urban):")
print(max_trans_urban)
```

```python
# checking the area with minimum transgender population both in rural and urban areas
min_trans_rural = df.loc[df['TRANSGENDER (RURAL)'].idxmin()]
min_trans_urban = df.loc[df['TRANSGENDER (URBAN)'].idxmin()]
print("Min Transgender (Rural):")
print(min_trans_rural)
print("\nmin Transgender (Urban):")
print(min_trans_urban)
```
