```python
#find sum of All sexes (Rural)
df['ALL SEXES (RURAL)'].sum()
```

```python
# find sum of All sexes (Rural) in 1998
df['POPULATION 1998 (RURAL)'].sum()
```

```python
# how much rural population increased from 1998 to 2017
df['ALL SEXES (RURAL)'].sum() - df['POPULATION 1998 (RURAL)'].sum()
```

```python
# calculate percentage increase
increase = (df['ALL SEXES (RURAL)'].sum() - df['POPULATION 1998 (RURAL)'].sum()) / df['POPULATION 1998 (RURAL)'].sum() * 100
increase
```

```python
# now find the population increase for urban areas as well
df['ALL SEXES (URBAN)'].sum() - df['POPULATION 1998 (URBAN)'].sum()
```

```python
# percentage increase in urban areas
increase_urban = (df['ALL SEXES (URBAN)'].sum() - df['POPULATION 1998 (URBAN)'].sum()) / df['POPULATION 1998 (URBAN)'].sum() * 100
increase_urban
```
